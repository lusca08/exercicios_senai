## Exemplo 4: Fun√ß√£o com c√°lculo

```python
def calcular_media(nota1, nota2, nota3):
    media = (nota1 + nota2 + nota3) / 3
    return media
```

```python
media_final = calcular_media(7, 8.5, 9)
print(f"Sua m√©dia foi: {media_final}")
```
